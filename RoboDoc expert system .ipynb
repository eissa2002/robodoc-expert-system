```python
import tkinter as tk
from tkinter import ttk

categories = {
          'children': {
        'hyper active': ['KINDIVAL'],
        'bedwetting': ['ENUKIND'],
        'cold': ['NISIKIND'],
        'growth deficiency': ['ANEKIND'],
        'fever': ['ECHINACEA ANGUSTIFOLIA 1X'],
        'calcium deficiency': ['CALCIOKIND'],
        'cough': ['TUSSIKIND'],
    },
    'digestive': {
        'diarrhoea': ['BIOCOMBINATION NO. 08'],
        'vomitings': ['ALPHA-MS'],
        'acidity': ['ALPHA-ACID'],
        'colic': ['COLIKIND'],
        'nausea': ['KINDIGEST'],
        'motion sickness': ['ALPHA-MS'],
        'loss of appetite': ['ALPHA-LIV'],
        'travel sickness': ['ALPHA-MS'],
        'constipation': ['NATRUM MURIATICUM'],
        'dysentery': ['HOLARRHENA ANTIDYSENTERICA 1X'],
    },
    'fever and general': {
        'fever': ['ACONITUM PENTARKAN'],
        'immune boosters': ['MUNOSTIM'],
        'tonsillitis': ['ALPHA-TONS'],
        'vomiting sensation': ['ALPHA-LIV'],
        'chill': ['ALPHA-WD'],
        'cold': ['NISIKIND'],
        'cough': ['TUSSIKIND'],
    },
    'joints and muscles': {
        'cramps': ['MAGNESIUM PHOSPHORICUM'],
        'muscular pain': ['Alpha-MP'],
        'rheumatic pain': ['Alpha-MP'],
        'bone problems': ['Calcarea phosphorica'],
        'body ache': ['TOPI ARNICA CREAM'],
        'joints': ['Topi MP Gel'],
        'inflammation': ['Topi Heal Cream'],
        'muscles': ['Kali phosphoricum'],
        'spasms': ['MagnesiumPhosphoricum Pentarkan'],
        'osteoarthritis': ['Biocombination No. 19'],
    },
    'respiratory diseases': {
        'chronic cough': ['Alpha-CC'],
        'spasmodic cough': ['Alpha-Coff'],
        'cough syrup': ['Aconitum Pentarkan'],
        'irritating cough': ['Glycyrrhiza glabra 1X'],
        'nasal congestion': ['Alpha-NC'],
        'sneezing': ['Alpha-RC'],
        'stuffiness of nostrils': ['Alpha-NC2'],
        'breathlessness': ['Alpha-RC1'],
        'coryza': ['BIOCOMBINATION NO. 05'],
        'respiratory congestion': ['Biocombination No. 13'],
    },
    'skin diseases': {
            'acne': ['TOPI BERBERIS CREAM'],
            'pimples': ['Calcarea sulphurica'],
            'psoriasis': ['Alpha-WD'],
            'bed sores': ['Topi Heal Cream'],
            'boils': ['Silicea'],
            'cuts': ['Hepar Sulphuris Pentarkan'],
            'itching': ['Graphites Pentarkan'],
            'scabies': ['Topi Sulphur Cream'],
            'dermatities': ['B&T Akne - Sor Soap'],
            'fungal infection': ['Thuja Occidentalis 1X'],
            'skin allergies': ['Azadirachta indica 1X'],
        },
}

class HomeopathicExpertSystemGUI:
    def __init__(self, master):
        self.master = master
        master.title("RoboDoc Expert System")

        # Create the welcome message label
        self.welcome_label = ttk.Label(master, text="Welcome to the RoboDoc Expert System!")
        self.welcome_label.grid(row=0, column=0, columnspan=2, padx=10, pady=10)

        # Create the category selection label and dropdown menu
        self.category_label = ttk.Label(master, text="Select a category:")
        self.category_label.grid(row=1, column=0, padx=10, pady=10)
        self.category_var = tk.StringVar()
        self.category_dropdown = ttk.Combobox(master, textvariable=self.category_var, values=list(categories.keys()))
        self.category_dropdown.grid(row=1, column=1, padx=10, pady=10)

        # Create the problem selection label and dropdown menu
        self.problem_label = ttk.Label(master, text="Select a problem:")
        self.problem_label.grid(row=2, column=0, padx=10, pady=10)
        self.problem_var = tk.StringVar()
        self.problem_dropdown = ttk.Combobox(master, textvariable=self.problem_var, values=[])
        self.problem_dropdown.grid(row=2, column=1, padx=10, pady=10)

        # Create the submit button
        self.submit_button = ttk.Button(master, text="Submit", command=self.submit)
        self.submit_button.grid(row=3, column=0, columnspan=2, padx=10, pady=10)

        # Create the medicine listbox
        self.medicine_listbox = tk.Listbox(master)
        self.medicine_listbox.grid(row=4, column=0, columnspan=2, padx=10, pady=10)

    def submit(self):
        # Get the selected category and problem
        category = self.category_var.get()
        problem = self.problem_var.get()

        # Get the associated medicines for the chosen problem
        medicines = categories[category][problem]

        # Clear the medicine listbox and populate it with the medicines
        self.medicine_listbox.delete(0, tk.END)
        for medicine in medicines:
            self.medicine_listbox.insert(tk.END, medicine)

# Create the main Tkinter window and the HomeopathicExpertSystemGUI instance
root = tk.Tk()
gui = HomeopathicExpertSystemGUI(root)

# Populate the problem dropdown menu based on the selected category
def update_problem_dropdown(*args):
    category = gui.category_var.get()
    problems = list(categories[category].keys())
    gui.problem_dropdown.config(values=problems)
gui.category_var.trace('w', update_problem_dropdown)

# Start the main event loop
root.mainloop()
```
